### Obtencion de la funcion de transferencia

Defino el nodo $V_x$ aplicando un divisor de tension:

$\displaystyle V_x=\frac{R_3}{R_3+\frac{1}{SC_1}}\cdot V_1 \quad\to\quad \displaystyle V_x=\frac{SC_1R_3}{SC_1R_3+1}\cdot V_1\quad (1)$

Igualo las corrientes que circulan por $R_1$ y $R_2$

$I_{R_1} = I_{R_2} \quad\to\quad \displaystyle\frac{V_1-V_x}{R_1}=\frac{V_x-V_2}{R_2} \quad\to\quad
\frac{V_1}{R_1}+\frac{V_2}{R_2}=\frac{V_x}{R_2}+\frac{V_x}{R_1} \quad\to\quad
\displaystyle\frac{V_1}{R_1}+\frac{V_2}{R_2}=V_x\cdot\frac{R_1+R_2}{R_1\cdot R_2}\quad (2)$

Reemplazo la ecuacion $(1)$ en la $(2)$:

$\displaystyle\frac{V_1}{R_1}+\frac{V_2}{R_2}=\left(\frac{SC_1R_3}{SC_1R_3+1}\cdot V_1\right)\cdot\frac{R_1+R_2}{R_1\cdot R_2}
\quad\to\quad\displaystyle\frac{V_1}{R_1}+\frac{V_2}{R_2}=V_1\cdot\displaystyle\left(\frac{SR_3R_1C_1+SR_3R_2C_1}{SR_3R_2R_1C_1+R_1R_2}\right)$

$V_1\cdot\left ( \displaystyle\frac{1}{R_1}-\displaystyle\frac{SR_3R_1C_1+SR_3R_2C_1}{SR_3R_2R_1C_1+R_2R_1}\right )
=-\displaystyle\frac{V_2}{R_2} \quad\to\quad
V_1\cdot\displaystyle\frac{SR_3R_2R_1C_1+R_2R_1-SR_3R_1^{2}C_1-SR_3R_2R_1C_1}
{R_1(SR_3R_2R_1C_1+R_2R_1)}=-\frac{V_2}{R_2}$

$\displaystyle\frac{R_2^2-SR_3R_2R_1C_1}{R_1(SR_3R_2C_1+R_2)}=-\frac{V_2}{V_1} \quad\to\quad
T(S)=\displaystyle\frac{V_2}{V_1}=\displaystyle\frac{SR_3R_2R_1C_1-R_2^2}{R_1(SR_3R_2C_1+R_2)}=
\displaystyle\frac{R_3R_2R_1C_1}{R_3R_2R_1C_1}\cdot
\displaystyle\frac{S-\displaystyle\frac{R_2^{2}}{R_3R_2R_1C_1}}
{S+\displaystyle\frac{R_2}{R_3R_2C_1}}$

$T(S)=\displaystyle\frac{S-\frac{R_2}{R_3R_1C_1}}{S+\frac{1}{R_3C_1}} \quad;\quad\quad$
Para un ratio unitario: $\quad\quad\displaystyle\frac{R_2}{R_1}=1 \quad\to\quad R_2=R_1\quad$ Reemplazo en $T(s)$

$T(S)=\displaystyle\frac{S-\frac{R_1}{R_3R_1C_1}}{S+\frac{1}{R_3C_1}}\quad\to\quad
T(S)=\displaystyle\frac{S-\frac{1}{R_3C_1}}{S+\frac{1}{R_3C_1}}$

A partir de la expresion anterior obtengo el modulo y fase de la transferencia:

$T(S=j\omega)=\displaystyle\frac{j\omega-\frac{1}{R_3C_1}}{j\omega+\frac{1}{R_3C_1}}\quad;\qquad
\left| T(\omega) \right| = \frac{ \sqrt{ \omega^2 + \left( \frac{1}{R_3 C_1} \right)^2 } }
{ \sqrt{ \omega^2 + \left( \frac{1}{R_3 C_1} \right)^2 } } = 1$

$\gamma(\omega)=arctg\left(\displaystyle\frac{\omega}{-\frac{1}{R_3 C_1}} \right) - 
arctg\left( \displaystyle\frac{\omega}{\frac{1}{R_3 C_1}} \right)$

```python
# Librerías externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np

# Librería de TC2, esta la vas a usar mucho
from pytc2.sistemas_lineales import bodePlot

qq = np.sqrt(2)/2

my_tf = TransferFunction( [1,-1], [1, 1] )

plt.close('all')

bodePlot(my_tf, fig_id=2, filter_description = 'Q={:3.3f}'.format(qq) )

fig = plt.figure(2)
axs = fig.get_axes()
axs[0].set_ylim([-40, 20]) 
```

### Deduccion del tipo de filtro que es esta red circuital

Debido a los graficos presentados y al diagrama de polos y ceros podemos identificar que este  filtro es un cambiador de fase pero no de amplitud.
Vemos que la respuesta de fase para frecuencias muy bajas es de 180° y que con el diagrama de polos y ceros usando un vector que apunta desde el polo o el cero al eje imaginario a 45°, se restan los angulos 135° - 45° y se obtiene 90° en la respuesta de fase.
Deducimos que los catetos del triangulo que forma el vector con los ejes deben ser iguales para formar 45°, entonces si la frecuencia angular cambia de fase a 90° debe ser para $\omega=\frac{1}{R_3C_1}$. Y por ultimo, vemos que para altas frecuencias la respuesta de fase tiende a no cambiar y mantenerse en 0°.

### Normalizacion de la transferencia

Transferencia a normalizar:

$T(S)=\displaystyle\frac{S-\frac{1}{R_3C_1}}{S+\frac{1}{R_3C_1}}$

Aplico la siguiente normalizacion y factor de escala:

$\left\{\begin{matrix}
C'=C\cdot\displaystyle\frac{\Omega_\omega}{\Omega_Z} \\
R'=\displaystyle\frac{R}{\Omega_Z}\\
S'=\displaystyle\frac{S}{\Omega_\omega}
\end{matrix}\right.
\qquad
\left\{\begin{matrix}
\Omega_\omega=\displaystyle\frac{1}{R_3C_1}\\
\Omega_Z=R_3
\end{matrix}\right.
$

$C'=C\cdot\displaystyle\frac{\Omega_\omega}{\Omega_Z}=
C\cdot\displaystyle\frac{1}{R_3C_1}\cdot R_3=1$

$R_3'= \displaystyle\frac{R_3}{\Omega_Z}=\frac{R_3}{R_3}=1 \quad$ 
y $\quad S'=\displaystyle\frac{S}{\Omega_\omega}=\frac{1}{1/(R_3C_1)} $

Finalmente obtengo la normalizacion de la transferencia

$T=\displaystyle\frac{S-\frac{1}{R_3C_1}}{S+\frac{1}{R_3C_1}}=
\frac{\frac{S'}{R_3C_1}-\frac{1}{R_3C_1}} {\frac{S'}{R_3C_1}-\frac{1}{R_3C_1}}=
\frac{S'-1}{S'+1}$
